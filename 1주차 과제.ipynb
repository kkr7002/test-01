
언어 분야 관련 기술은 구글 번역기 등이 있다, 구글 번역기는 인공지능을 이용해 반복학습을 통해 경험을 축적시켰고 보다 정확한 번역이 가능해졌다. 
음성 분야 관련 기술은 애플의 음성 기반 개인비서 서비스인 시리가 있다. 시리는 아이폰 사용자의 음성명령을 바탕으로 모바일 검색은 물론, 일정관리, 전화 걸기, 메모, 음악 재생 등 다양한 생활편의 서비스를 제공하는 개인비서 서비스다.  
이미지 관련 기술은 자동 채색 및 펜선 따기등이 개발되고 있다. 자동채색은 인공지능이 학습을 계속해 그림에 맞는 색을 자동으로 칠해주는 서비스이다. AI 딥러닝 기술을 이용해 대강 그린 스케치 그림에서 명확한 펜 선을 만들어준다.
자율주행 관련 기술은 독일 자동차업체 BMW가 내놓은 기술이 인상적이다. BMW는 전기차 ‘i3’에 자동주차 기술을 탑재했다. BMW의 자동주차 기술에는 충돌 회피 기술이 적용됐다. 차량에 장착된 4개의 레이저 스캐너가 주변 환경을 탐지하고 자동차가 장애물과 충돌하지 않도록 하는 기술이다.
